```python
##accuracy,reacll,percision
def comparison(y_train,y_test,y_pred_train,y_pred_test):
    from sklearn.metrics import recall_score , accuracy_score , precision_score
    acc_train=accuracy_score(y_train,y_pred_train)
    acc_test=accuracy_score(y_test,y_pred_test)
    r=recall_score(y_test,y_pred_test,average='macro')
    p=precision_score(y_test,y_pred_test,average='macro')
    print(f'acc_train={acc_train},acc_test={acc_test},recall={r},percision={p}')
    return acc_train,acc_test,r,p

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

#lily flower
ir = load_iris()
x=ir.data
y=ir.target
#preprocessing
from sklearn.preprocessing import StandardScaler
model=StandardScaler()
x=model.fit_transform(x)
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

#KNN
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)

#comparison and show with matplotlib
acc_train_knn,acc_test_Knn,r_Knn,p_Knn=comparison(y_train,y_test,y_pred_train,y_pred_test)
conparisonn = [acc_train_knn,acc_test_Knn,r_Knn,p_Knn]
title=['acc_train','acc_test','recall','percision']
colors=['red','green','blue','black']
plt.bar(title,conparisonn,color=colors)
plt.grid()
plt.show()
```
